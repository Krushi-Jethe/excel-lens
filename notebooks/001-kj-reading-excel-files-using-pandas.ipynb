```python
tables = []
tables_dict = {}
table_count = 1

# Finding multiple tables seperated by NaN rows
for idx, row in df.iterrows():
    if (idx == 0) and all(row.isnull()):
        continue

    if all(row.isnull()):
        if tables_dict == {}:
            tables_dict[f"table_{table_count}"] = df.iloc[:idx, :]
            curr_table_idx = idx
        else:
            if len(df.iloc[curr_table_idx:idx, :]) == 1 and all(
                df.iloc[curr_table_idx:idx, :].isnull()
            ):
                curr_table_idx = idx
                continue
            tables_dict[f"table_{table_count+1}"] = df.iloc[curr_table_idx:idx, :]
            curr_table_idx = idx
            table_count += 1

    if idx == len(df) - 1:
        tables_dict[f"table_{table_count+1}"] = df.iloc[curr_table_idx:, :]

# Dropping NaN columns which seperate tables
for table_num, table in tables_dict.items():

    unnamed_cols = [col for col in table.columns if "Unnamed" in col]
    drop_col_flag = False
    drop_cols_list = []

    if not unnamed_cols:
        continue
    for col in unnamed_cols:
        if all(table[col].isnull()):
            drop_cols_list.append(col)
    table.drop(columns=drop_cols_list, axis=1, inplace=True)
```

### Analyzing sheet: Calcs

```python
tables = []
tables_dict = {}
table_count = 1

# Finding multiple tables seperated by NaN rows
for idx, row in df.iterrows():
    if (idx == 0) and all(row.isnull()):
        continue

    if all(row.isnull()):
        if tables_dict == {}:
            tables_dict[f"table_{table_count}"] = df.iloc[:idx, :]
            curr_table_idx = idx
        else:
            if len(df.iloc[curr_table_idx:idx, :]) == 1 and all(
                df.iloc[curr_table_idx:idx, :].isnull()
            ):
                curr_table_idx = idx
                continue
            tables_dict[f"table_{table_count+1}"] = df.iloc[curr_table_idx:idx, :]
            curr_table_idx = idx
            table_count += 1

    if idx == len(df) - 1:
        tables_dict[f"table_{table_count+1}"] = df.iloc[curr_table_idx:, :]

# Dropping NaN columns which seperate tables
for table_num, table in tables_dict.items():

    unnamed_cols = [col for col in table.columns if "Unnamed" in col]
    drop_col_flag = False
    drop_cols_list = []

    if not unnamed_cols:
        continue
    for col in unnamed_cols:
        if all(table[col].isnull()):
            drop_cols_list.append(col)
    table.drop(columns=drop_cols_list, axis=1, inplace=True)


# Drop all NaN columns
for table_num, table in tables_dict.items():
    drop_cols_list = []
    for col in table.columns:
        if all(table[col].isnull()):
            drop_cols_list.append(col)
    table.drop(columns=drop_cols_list, axis=1, inplace=True)

for table_num, table in tables_dict.items():
    for idx, row in table.iterrows():
        if all(row.isnull()):
            tables_dict[table_num] = table.drop(index=idx, axis=0).reset_index(
                drop=True
            )
```
